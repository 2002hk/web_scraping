```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.coursera.org/search?query=data%20science&topic=Data%20Science&sortBy=BEST_MATCH',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
course_name=[]
for i in soup.find_all('h3',class_='cds-CommonCard-title'):
  #print(i.text.strip())
  course_name.append(i.text.strip())
print(course_name)
```

```python
institute=[]
for i in soup.find_all('p',class_='cds-ProductCard-partnerNames'):
  #print(i.text.strip())
  institute.append(i.text.strip())
print(institute)
```

```python
skills=[]
for i in soup.find_all('p',class_='css-vac8rf'):
  #print(i.text.strip())
  skills.append(i.text.strip())
print(skills)
```
